```python
import random
# import json

words = {
    "Animals" : ['lion','tiger','monkey','cat','dog','elephant','donkey','hippopotamus','cow','goat','frog'],
    "Colours" : ['black','brown','yellow','orange','red','green','white','pink','blue','purple','golden','silver','grey'],
    "Fruits" : ['orange','mango','strawberry','mulberry','raspberry','plum','cherry','jackfruit','watermelon','muskmelon','apple','jamun']
}

hangman_stages = [
    '''
       +---+
       |   |
           |
           |
           |
           |
    =========''',
    '''
       +---+
       |   |
       O   |
           |
           |
           |
    =========''',
    '''
       +---+
       |   |
       O   |
       |   |
           |
           |
    =========''',
    # '''
    #    +---+
    #    |   |
    #    O   |
    #   /|   |
    #        |
    #        |
    # =========''',
    '''
       +---+
       |   |
       O   |
      /|\\  |
           |
           |
    =========''',
    '''
       +---+
       |   |
       O   |
      /|\\  |
      /    |
           |
    =========''',
    '''
       +---+
       |   |
       O   |
      /|\\  |
      / \\  |
           |
    ========='''
]

category = random.choice(list(words.keys()))
word = random.choice(words[category])
print('Category = ',category)

letters = set(word)
guessed_letters = set()
wrong_attempts = 0
max_attempts = len(hangman_stages) - 1

while len(letters) > 0 and wrong_attempts < max_attempts:
    print(hangman_stages[wrong_attempts])
    display = [i if i in guessed_letters else '_' for i in word]
    print("Current word = ",' '.join(display))
    print("Attempts left :",max_attempts - wrong_attempts)
    print()

    while True:
        guess = input('Guess a letter: ').lower()
        if len(guess) == 1 and guess.isalpha():
            break
        else:
            print('Enter a single letter.')

    if guess in letters:
        guessed_letters.add(guess)
        letters.remove(guess)
    elif guess in guessed_letters:
        print("Already guessed.")
    else:
        guessed_letters.add(guess)
        wrong_attempts += 1
        # max_attempts -= 1
        print("Already guessed letters : ",guessed_letters)
        print("Wrong answer.")

if len(letters) == 0:
    print(hangman_stages[wrong_attempts])
    print(f'Congratulations! You guessed the word: {word}')
else:
    print(hangman_stages[wrong_attempts])
    print("Attempts left : 0")
    print(f'Sorry, you lost. The word was: {word}')
```
